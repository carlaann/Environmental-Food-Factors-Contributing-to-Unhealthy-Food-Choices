# Cumulative Distrubution Function

```python
#Using Food Tax to show how the tax on food is distrubuted across the data
```

```python
################# cdf #########################

def EvalCdf(sample, x):
    count = 0.0
    for value in sample:
        if value <= x:
            count += 1

    prob = count / len(sample)
    return prob
```

```python
EvalCdf(df_tax.FOOD_TAX14,1)
```

# Analytical Distribution 

```python
def MakeNormalPlot(weights):
    mean = weights.mean()
    std = weights.std()
    xs = -2, 2
    fxs, fys = thinkstats2.FitLine(xs, inter=mean, slope=std)
    thinkplot.Plot(fxs, fys, color='gray', label='Fitted Line')
    xs,ys = thinkstats2.NormalProbability(weights)
    thinkplot.Plot(xs,ys, label='All Food Taxes')
    thinkplot.Config(xlabel='standard deviation from the mean',
                 ylabel='food taxes')
    
```

```python
MakeNormalPlot(df_merge_subset.FOOD_TAX14)
```

```python
# Normal probality shows there is no normal distribution of taxes within states and counties.  
```

```python
import random
```

```python
class CoinTest(thinkstats2.HypothesisTest):
    def TestStatistic(self, data):
        heads, tails = data
        test_stat = abs(heads - tails)
        return test_stat
    def RunModel(self):
        heads, tails = self.data
        n = heads + tails
        sample = [random.choice('HT') for _ in range(n)]
        hist = thinkstats2.Hist(sample)
        data = hist['H'], hist['T']
        return data
```

```python
ct = CoinTest((2147, 996))
pvalue = ct.PValue()
```

```python
pvalue
```

```python
# p-value of 0 indicates a failure to reject the null hypothesis at the 5% significance level.
```

```python
ct.TestStatistic((2147, 996))
```

```python
ct.RunModel()
```
